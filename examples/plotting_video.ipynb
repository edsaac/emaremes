# Control map over time

```python
from pathlib import Path
from functools import cache
import matplotlib.pyplot as plt

import emaremes as mrms
from ipywidgets import SelectionSlider, interact
```

```python
mrms.fetch.path_config.set_prefered("./data")

prate_files = mrms.fetch.timerange(
    "20240927-060000", "20240927-160000", frequency="1h", data_type="precip_rate", verbose=True
)
```

```python
@cache
def make_frame(f: Path) -> plt.Figure:
    fig = mrms.plot.precip_rate_map(f, state="NC")
    return fig
```

```python
selected_file = SelectionSlider(options=prate_files, value=prate_files[0], description="File")
interact(make_frame, f=selected_file)
```
