```python
from pathlib import Path
import geopandas as gpd
from emaremes.timeseries import extract_multipolygon_series
```

```python
# Read polygons
geodata = gpd.read_file("../data_old/Subcatchments__WestLittleCal/Subcatchments__WestLittleCal.shp")
```

```python
# Select a GRIB2 file
data_folder = Path("../data_old/20210626/")

# Clean up old IDX
idx_files = data_folder.glob("*.idx")
for idx in idx_files:
    idx.unlink()

# Find GRIB2 files
files = data_folder.glob("*.grib2")
files = sorted(files)
print(len(files))
```

```python
df = extract_multipolygon_series(files[350:400], geodata, upsample=True)
```
