```python
import xarray as xr
from pathlib import Path
```

```python
data_folder = Path("../data_old/20210626/")
grib_files = data_folder.glob("*.grib2")
grib_files = sorted(grib_files)
```

```python
ds = xr.open_mfdataset(
    grib_files, engine="cfgrib", combine="nested", concat_dim="time", decode_timedelta=False
)
```

```python
ds.sel(latitude=40, longitude=270, method="nearest")["unknown"].plot()
```
