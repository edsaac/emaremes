```python
import gzip
from pathlib import Path
from io import BytesIO
from tempfile import NamedTemporaryFile

import xarray as xr
```

```python
gzip_files = Path("../data_old/20210628").glob("*.grib2.gz")
gzip_files = sorted(gzip_files)
gzip_file = gzip_files[0]
print(gzip_file)

grib_files = Path("../data_old/20210628").glob("*.grib2")
grib_files = sorted(grib_files)
grib_file = grib_files[0]
print(grib_file)
```

```python
with xr.open_dataset(grib_file, engine="cfgrib") as xar:
    display(xar)
```

```python
with gzip.open(gzip_file, "rb") as gzin:
    with NamedTemporaryFile("ab+", suffix=".grib2") as tf:
        unzipbytes = gzin.read()
        tf.write(unzipbytes)

        with xr.open_dataset(tf.name, engine="cfgrib") as xar:
            display(xar)
```
